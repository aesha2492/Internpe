```python
import pandas as pd
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise.accuracy import rmse

# Step 1: Load the dataset
data = Dataset.load_builtin('ml-100k')

# Step 2: Create train and test sets
trainset, testset = train_test_split(data, test_size=0.2)

# Step 3: Train the model
model = SVD()
model.fit(trainset)

# Step 4: Evaluate the model
predictions = model.test(testset)
accuracy = rmse(predictions)
print("RMSE:", accuracy)

# Step 5: Generate recommendations
user_id = str(196)  # Example user ID

# Get all item IDs
item_ids = set()
for item in trainset.all_items():
    item_ids.add(item)

# Remove the items already rated by the user
user_items = trainset.ur[int(user_id)]
user_unseen_items = [item for item in item_ids if item not in user_items]

# Predict ratings for unseen items
predictions = []
for item_id in user_unseen_items:
    prediction = model.predict(user_id, item_id)
    predictions.append((item_id, prediction.est))

# Step 6: Sort and display recommendations
recommendations = sorted(predictions, key=lambda x: x[1], reverse=True)[:10]
for item_id, rating in recommendations:
    print(f"Item ID: {item_id}, Predicted Rating: {rating}")
```
